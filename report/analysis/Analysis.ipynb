```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Benchmark
num2 = pd.read_excel('RawData/CPUUsage.xlsx', parse_cols='A')
cu = pd.read_excel('RawData/CPUUSage.xlsx', parse_cols='B')
plt.plot(num2, cu)
plt.ylabel('CPU Usage (%)')
plt.xlabel('Number of Clients')
plt.show()

num1 = pd.read_excel('RawData/MemoryUsage.xlsx', parse_cols='A')
mu = pd.read_excel('RawData/MemoryUsage.xlsx', parse_cols='B')
plt.plot(num1, mu)
plt.ylabel('Memory Usage (MB)')
plt.xlabel('Number of Clients')
plt.show()
```

```python
# Responsiveness
resp = pd.read_csv('RawData/responsiveness.csv')
ax = resp.plot(kind='line', x='Seconds')
plt.ylabel('Average Response Time (Milliseconds)')
plt.xlabel('Time (Seconds)')
plt.show()
```

```python
# Consistency Deviation
resp = pd.read_csv('RawData/consistency-dev.csv')
ax = resp.plot(kind='line', x='Seconds', legend=False)
plt.ylabel('Deviation (Board Squares)')
plt.xlabel('Transition Points')
plt.show()
```

```python
# Fault Tolerance - Responsiveness
resp = pd.read_csv('RawData/fault-tolerance-2.csv')
ax = resp.plot(kind='line', x='Transition')
plt.ylabel('Response Time (Milliseconds)')
plt.xlabel('Transition Points')
ax.set_xticklabels(['', '2-1', '', '3-2', '', '4-3', '', '5-4'])
plt.gca().invert_xaxis()
plt.show()
```
